```python
#Load the data setin float value
df = pd.read_excel('oil well 1.xlsx')
df.columns = df.columns.str.strip() # Remove leading/trailing spaces from column names
df['water_cut_float'] = df['Water cut (%)'].astype(str).str.replace('%', '').astype(float) / 100
```

```python
df
```

```python
#LINE PLOT
plt.plot(df['\nReservoir pressure (atm)'],df['\nOil volume (m3/day)'])
plt.title('Oil volume vs Reservoir pressure')
plt.xlabel('Reservoir pressure (atm)')
plt.ylabel('Oil volume (m3/day)')
plt.show()

```

```python
#LINE PLOT
plt.plot(df['Reservoir pressure (atm)'],df['water_cut_float'])
plt.title('Water cut vs Reservoir pressure')
plt.xlabel('Reservoir pressure (atm)')
plt.ylabel('Water cut')
plt.show()
```

```python
#LINE PLOT
plt.plot(df['\nReservoir pressure (atm)'],df['Water volume (m3/day)'])
plt.title('Water volume vs Reservoir pressure')
plt.xlabel('Reservoir pressure (atm)')
plt.ylabel('Water volume (m3/day)')
plt.show()
```

```python
#BAR PLOT
plt.bar(df['\nDynamic level (m)'],df['\nGas volume (m3/day)'])
plt.title('Gas volume vs Dynamic level')
plt.xlabel('Dynamic level (m)')
plt.ylabel('Gas volume (m3/day)')
plt.show()
```

```python
#BAR PLOT
plt.bar(df['\nDynamic level (m)'],df['\nOil volume (m3/day)'])
plt.title('Oil volume vs Dynamic level')
plt.xlabel('Dynamic level (m)')
plt.ylabel('Oil volume (m3/day)')
plt.show()
```

```python
#scatter plot
plt.scatter(df['Dynamic level (m)'],df['Gas volume (m3/day)'])
plt.title('Gas volume vs Dynamic level')
plt.xlabel('Dynamic level (m)')
plt.ylabel('Gas volume (m3/day)')
plt.show()
```

```python
plt.scatter(df['Dynamic level (m)'],df['Reservoir pressure (atm)'])
plt.title('Reservoir pressure vs Dynamic level')
plt.xlabel('Dynamic level (m)')
plt.ylabel('Reservoir pressure (atm)')
plt.show()
```

```python
plt.scatter(df['Reservoir pressure (atm)'],df['water_cut_float'])
plt.title('Reservoir pressure vs Water cut')
plt.xlabel('Reservoir pressure')
plt.ylabel('Water cut')
plt.show()
```

```python
#line plot
plt.scatter(df['Working hours'],df['Oil volume (m3/day)'])
plt.title('Oil volume vs Working hours')
plt.xlabel('Working hours')
plt.ylabel('Oil volume (m3/day)')
plt.show()
```

```python
#subplots
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].scatter(df['Reservoir pressure (atm)'],df['Oil volume (m3/day)'])
ax[0].set_title('Reservoir pressure (atm) vs Oil volume ')
ax[0].set_xlabel('Reservoir pressure (atm)')
ax[0].set_ylabel('Oil volume (m3/day)')
ax[1].scatter(df['Reservoir pressure (atm)'],df['Gas volume (m3/day)'])
ax[1].set_title('Reservoir pressure (atm) vs Gas volume')
ax[1].set_xlabel('Reservoir pressure (atm)')
ax[1].set_ylabel('Gas volume (m3/day)')
plt.show()
```

```python
#Histogram plot
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].hist(df['Oil volume (m3/day)'],bins=20)
ax[0].set_title('Oil volume (m3/day)')
ax[0].set_xlabel('Oil volume (m3/day)')
ax[0].set_ylabel('Frequency')
ax[1].hist(df['Gas volume (m3/day)'],bins=20)
ax[1].set_title('Gas volume (m3/day)')
ax[1].set_xlabel('Gas volume (m3/day)')
ax[1].set_ylabel('Frequency')
plt.show()
```

```python
#subplots
fig, ax = plt.subplots(2,2,figsize=(15,15))
ax[0,0].scatter(df['Reservoir pressure (atm)'],df['Oil volume (m3/day)'])
ax[0,0].set_title('Reservoir pressure (atm) vs Oil volume ')
ax[0,0].set_xlabel('Reservoir pressure (atm)')
ax[0,0].set_ylabel('Oil volume (m3/day)')
ax[0,1].scatter(df['Reservoir pressure (atm)'],df['Gas volume (m3/day)'])
ax[0,1].set_title('Reservoir pressure (atm) vs Gas volume')
ax[0,1].set_xlabel('Reservoir pressure (atm)')
ax[0,1].set_ylabel('Gas volume (m3/day)')
ax[1,0].scatter(df['Reservoir pressure (atm)'],df['Volume of liquid (m3/day)'])
ax[1,0].set_title('Reservoir pressure (atm) vs Volume of liquid')
ax[1,0].set_xlabel('Reservoir pressure (atm)')
ax[1,0].set_ylabel('Volume of liquid (m3/day)')
ax[1,1].scatter(df['Reservoir pressure (atm)'],df['Water volume (m3/day)'])
ax[1,1].set_title('Reservoir pressure (atm) vs Water volume')
ax[1,1].set_xlabel('Reservoir pressure (atm)')
ax[1,1].set_ylabel('Water volume (m3/day)')
plt.show()



```
